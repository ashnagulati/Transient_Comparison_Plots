```python
import pandas as pd
import numpy as np

# Load the CSV
df_equi = pd.read_csv("E_gammabeta_equipartition.csv")

# Add/modify columns
df_equi["beta"] = np.nan * np.ones(len(df_equi))        # fill with NaNs
df_equi["Lorentz factor"] = np.nan * np.ones(len(df_equi))   
df_equi["Type"] = df_equi["Name"]
df_equi["Relativicity"] = ["Non-rel"] * len(df_equi)
df_equi["Ref"] = ["This work"] * len(df_equi)

# Rename column
df_equi.rename(columns={"E_center": "Energy"}, inplace=True)

# Show selected columns
df_equi_subset = df_equi[["Name", "beta", "Lorentz factor", "Gamma_beta", "Energy", "Type", "Relativicity","Ref"]]
df_equi_subset
```

```python
all_data=pd.concat([df_other_trans,df_equi_subset])
all_data
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# Define your custom styles
plot_style = {
    'GRBs':                     {'marker': 'o', 'alpha': 1, 'size': 70, 'zorder': 1, 'color': "tab:blue"},
    'Sub-E GRBs':               {'marker': 'o', 'alpha': 1, 'size': 70, 'zorder': 2, 'color': "lightblue"},
    'SNe':                      {'marker': 'o', 'alpha': 1, 'size': 70, 'zorder': 1, 'color': "dimgrey"},
    'LLAGN':                      {'marker': 'X', 'alpha': 1, 'size': 90, 'zorder': 1, 'color': "purple"},
    # 'Rel-Sne':        {'marker': 'o', 'alpha': 1, 'size': 40, 'zorder': 2, 'color': "darkgrey"},
    'TDEs (jetted)':            {'marker': 's', 'alpha': 1, 'size': 60, 'zorder': 7, 'color': "darkgreen"},
    'TDEs':                     {'marker': 's', 'alpha': 1, 'size': 60, 'zorder': 7, 'color': "yellowgreen"},
    'Delayed TDEs':             {'marker': 's', 'alpha': 1, 'size': 60, 'zorder': 7, 'color': "darkseagreen"},
    'FBOTs':                    {'marker': 'D', 'alpha': 1, 'size': 60, 'zorder': 50, 'color': "black"},
    'ASKAP J0055-2558_sph': {'marker': '*', 'alpha': 1, 'size': 300, 'zorder': 100, 'color': "peru"},
    'ASKAP J0055-2558_c15': {'marker': '*', 'alpha': 1, 'size': 300, 'zorder': 100, 'color': "gold"},
    'ASKAP J0055-2558_c30': {'marker': '*', 'alpha': 1, 'size': 300, 'zorder': 100, 'color': "orange"},
    'ASKAP J0055-2558_c5': {'marker': '*', 'alpha': 1, 'size': 300, 'zorder': 100, 'color': "red"},
    "OA": {'marker': '*', 'alpha': 0.6, 'size': 100, 'zorder': 100,'color':"black"}
    # 'ASKAP J0055-2558_rel_c15': {'marker': '*', 'alpha': 1, 'size': 100, 'zorder': 100, 'color': "red"},
    # "ASKAP J0055-2558_non_rel": {'marker': 's', 'alpha': 1, 'size': 50, 'zorder': 100, 'color': "red"},
}
special_types_with_boxes = [] #["ASKAP J0055-2558_rel", "ASKAP J0055-2558_non_rel"]

def plot_energy_vs_gamma_beta(all_data):
    fig = plt.figure(figsize=(7, 6))
    ax = plt.gca()

    used_labels = set()

    for (obj_name, obj_type), group in all_data.groupby(['Name', 'Type']):
        style = plot_style.get(obj_type, {'marker': 'o', 'alpha': 0.7, 'size': 30, 'zorder': 1, 'color': 'gray'})

        gamma_beta_list = []
        energy_list = []

        for idx, row in group.iterrows():
            gamma_beta_raw = str(row['Gamma_beta'])
            energy_raw = str(row['Energy'])

            if obj_type in special_types_with_boxes:
                # Special case: draw dashed rectangle
                gamma_beta_low, gamma_beta_high = parse_range(gamma_beta_raw)
                energy_low, energy_high = parse_range(energy_raw)

                width = gamma_beta_high - gamma_beta_low
                height = energy_high - energy_low

                rect = mpatches.Rectangle(
                    (gamma_beta_low, energy_low),
                    width, height,
                    linewidth=1,
                    linestyle='-',  # Dashed rectangle
                    edgecolor=style['color'],
                    facecolor='none',
                    zorder=style['zorder']
                )
                ax.add_patch(rect)

            else:
                gamma_beta_val, gamma_beta_err_low, gamma_beta_err_high = parse_value_with_errors(gamma_beta_raw)
                energy_val, energy_err_low, energy_err_high = parse_value_with_errors(energy_raw)

                gamma_beta_list.append(gamma_beta_val)
                energy_list.append(energy_val)

                gamma_beta_arrow = None
                energy_arrow = None
                if '<' in gamma_beta_raw:
                    gamma_beta_arrow = 'left'
                elif '>' in gamma_beta_raw:
                    gamma_beta_arrow = 'right'
                if '<' in energy_raw:
                    energy_arrow = 'down'
                elif '>' in energy_raw:
                    energy_arrow = 'up'

                if gamma_beta_arrow is None and energy_arrow is None:
                    xerr = np.array([
                        [abs(gamma_beta_val - gamma_beta_err_low)],
                        [abs(gamma_beta_err_high - gamma_beta_val)]
                    ])
                    yerr = np.array([
                        [abs(energy_val - energy_err_low)],
                        [abs(energy_err_high - energy_val)]
                    ])

                    # Draw errorbars without marker
                    ax.errorbar(
                        gamma_beta_val, energy_val,
                        xerr=xerr,
                        yerr=yerr,
                        fmt='none',  # no marker at center
                        ecolor=style['color'],
                        elinewidth=1.5,
                        capsize=3,
                        alpha=style['alpha'],
                        zorder=style['zorder'] 
                    )

                    # Now separately scatter the point
                    ax.scatter(
                        gamma_beta_val, energy_val,
                        marker=style['marker'],
                        color=style['color'],
                        s=style['size'],
                        alpha=style['alpha'],
                        label=obj_type if obj_type not in used_labels else None,
                        zorder=style['zorder'],
                        edgecolor="black"
                    )

                else:
                    # Just scatter + big arrow for limits
                    ax.scatter(
                        gamma_beta_val, energy_val,
                        marker=style['marker'],
                        color=style['color'],
                        s=style['size'],
                        alpha=style['alpha'],
                        label=obj_type if obj_type not in used_labels else None,
                        zorder=style['zorder'],
                        edgecolor="black"
                    )

                    # Draw bigger arrows
                    arrow_dx = 0.4 * gamma_beta_val  # Make arrow longer (40% of x)
                    arrow_dy = 2 * energy_val  # Arrow longer (70% of y)

                    if gamma_beta_arrow:
                        if gamma_beta_arrow == 'left':
                            ax.annotate('', xy=(gamma_beta_val - arrow_dx, energy_val), xytext=(gamma_beta_val, energy_val),
                                        arrowprops=dict(arrowstyle="->", color=style['color'], lw=2))
                        else:
                            ax.annotate('', xy=(gamma_beta_val + arrow_dx, energy_val), xytext=(gamma_beta_val, energy_val),
                                        arrowprops=dict(arrowstyle="->", color=style['color'], lw=2))

                    if energy_arrow:
                        if energy_arrow == 'down':
                            ax.annotate('', xy=(gamma_beta_val, energy_val - arrow_dy), xytext=(gamma_beta_val, energy_val),
                                        arrowprops=dict(arrowstyle="->", color=style['color'], lw=2))
                        else:
                            ax.annotate('', xy=(gamma_beta_val, energy_val + arrow_dy), xytext=(gamma_beta_val, energy_val),
                                        arrowprops=dict(arrowstyle="->", color=style['color'], lw=2))

            used_labels.add(obj_type)

        # After plotting all points for one object, connect them
        if len(gamma_beta_list) >= 2:
            gamma_beta_list, energy_list = zip(*sorted(zip(gamma_beta_list, energy_list)))
            ax.plot(
                gamma_beta_list, energy_list,
                color=style['color'],
                lw=1.5,
                alpha=style['alpha'],
                zorder=style['zorder'] - 2  # Line below scatter
            )

    # Draw vertical line at Gamma_beta = 1
    ax.axvline(x=1, color='black', linestyle='--', linewidth=2, zorder=0)

#     # Add labels for non-relativistic and relativistic regimes
    ax.text(0.35,0.95,r"Non-relativistic",horizontalalignment='left',verticalalignment='center', transform=ax.transAxes,color="black",fontsize="12",fontweight="bold")
    ax.text(0.65,0.95,r"Relativistic",horizontalalignment='left',verticalalignment='center', transform=ax.transAxes,color="black",fontsize="12",fontweight="bold")


    # Build the legend
    label_order = ['ASKAP J0055-2558_non_rel_sph','ASKAP J0055-2558_non_rel_c30',
                   'ASKAP J0055-2558_non_rel_c15','ASKAP J0055-2558_non_rel_c10',
                   'GRBs', 'Sub-E GRBs', 'SNe', 'FBOTs',
                   'TDEs (jetted)', 'TDEs', 'Delayed TDEs','LLAGN']

    handles = [
        mpatches.Patch(color=plot_style[label]['color'], label=label)
        for label in label_order
        if label in used_labels  # only if the label was actually used in the plot
    ]
    ax.legend(
        handles=handles,
        loc='upper center',
        bbox_to_anchor=(0.5, -0.15),
        ncol=3,
        frameon=False,
        fontsize=11,
        handletextpad=0.5,
        columnspacing=1.5
    )

    
    # Set log scales
    ax.set_xscale('log')
    ax.set_yscale('log')

    ax.set_xlabel(r'$\Gamma \beta$', fontsize=16)
    ax.set_ylabel(r'Kinetic Energy (erg)', fontsize=16)

    ax.tick_params(axis='both', colors='black', direction='in', length=7, width=1, labelsize="large")
    ax.tick_params(axis='both', which="minor", colors='black', direction='in', length=5, width=1)

    ax.set_yscale('log')

    from matplotlib.ticker import LogLocator
    ax.yaxis.set_minor_locator(LogLocator(base=10.0, subs=np.arange(2, 10)*0.1, numticks=10))
    ax.yaxis.set_minor_formatter(plt.NullFormatter())

    ax.tick_params(axis='y', which='both', direction='in', length=5, width=1)
    ax.tick_params(axis='both', which="major", colors='black', direction='in', length=7, width=1)


    for spine in ax.spines.values():
        spine.set_linewidth(2)

    plt.tight_layout()
    # plt.show()
    return fig

def parse_value_with_errors(value_str):
    value_str = value_str.strip()
    if '/' in value_str:
        parts = value_str.split('/')
        if len(parts)==2:
            center = float(parts[0])
            err_low = float(parts[1])
            err_u = float(parts[1])
        else:
            center = float(parts[0])
            err_low = float(parts[1])
            err_u = float(parts[2])
        return center, err_low, err_u
    elif '<' in value_str:
        val = float(value_str.replace('<', '').strip())
        return val, val * 0.5, val
    elif '>' in value_str:
        val = float(value_str.replace('>', '').strip())
        return val, val, val * 1.5
    else:
        val = float(value_str)
        return val, val, val

def parse_range(value_str):
    value_str = value_str.strip()
    if '-' in value_str:
        low, high = value_str.split('-')
        return float(low), float(high)
    else:
        raise ValueError(f"Expected a range with '-' but got {value_str}")
```

```python
def E_eq_to_E(E_eq, eps_e, eps_b, case):
    eps = (eps_b / eps_e) / (6 / 11)
    if case == "Newtonian":
        E = E_eq * (((11/17) * (eps ** (-6/17))) + ((6/17) * (eps ** (11/17))))
        return E
    else:
        raise ValueError("Only 'Newtonian' case is implemented.")

# Given parameters
E_eq = 1e51  # example E_eq value in erg
eps_e_values = [0.1, 0.0005]
eps_b_values = [0.1, 0.01, 0.001]

# Compute and collect combinations
results = []
for eps_e in eps_e_values:
    for eps_b in eps_b_values:
        E = E_eq_to_E(E_eq, eps_e, eps_b, case="Newtonian")
        results.append((E, eps_e, eps_b))

# Sort by energy E
results_sorted = sorted(results, key=lambda x: x[0])

# Display results
for E, eps_e, eps_b in results_sorted:
    print(f"E = {E:.3e} erg | eps_e = {eps_e} | eps_b = {eps_b}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd

# --- Your ASKAP data is assumed to be in all_data ---

plot_style = {
    'GRBs': {'marker': 'o', 'alpha': 1, 'size': 70, 'zorder': 1, 'color': "tab:blue"},
    'Sub-E GRBs': {'marker': 'o', 'alpha': 1, 'size': 70, 'zorder': 2, 'color': "lightblue"},
    'SNe': {'marker': 'o', 'alpha': 1, 'size': 70, 'zorder': 1, 'color': "dimgrey"},
    'LLAGN': {'marker': 'X', 'alpha': 1, 'size': 90, 'zorder': 1, 'color': "purple"},
    'TDEs (jetted)': {'marker': 's', 'alpha': 1, 'size': 60, 'zorder': 7, 'color': "darkgreen"},
    'TDEs': {'marker': 's', 'alpha': 1, 'size': 60, 'zorder': 7, 'color': "yellowgreen"},
    'Delayed TDEs': {'marker': 's', 'alpha': 1, 'size': 60, 'zorder': 7, 'color': "darkseagreen"},
    'FBOTs': {'marker': 'D', 'alpha': 1, 'size': 60, 'zorder': 50, 'color': "black"},
    'ASKAP J0055-2558_sph': {'marker': '*', 'alpha': 1, 'size': 300, 'zorder': 100, 'color': "peru"},
    'ASKAP J0055-2558_c15': {'marker': '*', 'alpha': 1, 'size': 300, 'zorder': 100, 'color': "gold"},
    'ASKAP J0055-2558_c30': {'marker': '*', 'alpha': 1, 'size': 300, 'zorder': 100, 'color': "orange"},
    'ASKAP J0055-2558_c5': {'marker': '*', 'alpha': 1, 'size': 300, 'zorder': 100, 'color': "red"},
    "OA": {'marker': '*', 'alpha': 0.6, 'size': 100, 'zorder': 100, 'color': "black"}
}

special_types_with_boxes = []

def parse_value_with_errors(value_str):
    value_str = value_str.strip()
    if '/' in value_str:
        parts = value_str.split('/')
        if len(parts) == 2:
            center = float(parts[0])
            err = float(parts[1])
            return center, center - err, center + err
        elif len(parts) == 3:
            center = float(parts[0])
            err_low = float(parts[1])
            err_u = float(parts[2])
            return center, center - err_low, center + err_u
    elif '<' in value_str:
        val = float(value_str.replace('<', '').strip())
        return val, val * 0.5, val
    elif '>' in value_str:
        val = float(value_str.replace('>', '').strip())
        return val, val, val * 1.5
    else:
        val = float(value_str)
        return val, val, val

def parse_range(value_str):
    value_str = value_str.strip()
    if '-' in value_str:
        low, high = value_str.split('-')
        return float(low), float(high)
    else:
        raise ValueError(f"Expected a range with '-' but got {value_str}")

def plot_energy_vs_gamma_beta(all_data):
    fig = plt.figure(figsize=(7, 6))
    ax = plt.gca()
    used_labels = set()

    for (obj_name, obj_type), group in all_data.groupby(['Name', 'Type']):
        style = plot_style.get(obj_type, {'marker': 'o', 'alpha': 0.7, 'size': 30, 'zorder': 1, 'color': 'gray'})
        gamma_beta_list = []
        energy_list = []

        for _, row in group.iterrows():
            gamma_beta_raw = str(row['Gamma_beta'])
            energy_raw = str(row['Energy'])

            if obj_type in special_types_with_boxes:
                continue  # skipping boxed range types for now
            else:
                gamma_beta_val, _, _ = parse_value_with_errors(gamma_beta_raw)
                energy_val, _, _ = parse_value_with_errors(energy_raw)

                gamma_beta_list.append(gamma_beta_val)
                energy_list.append(energy_val)

                ax.scatter(
                    gamma_beta_val, energy_val,
                    marker=style['marker'],
                    color=style['color'],
                    s=style['size'],
                    alpha=style['alpha'],
                    label=obj_type if obj_type not in used_labels else None,
                    zorder=style['zorder'],
                    edgecolor="black"
                )

            used_labels.add(obj_type)

        # Connect points per object
        if len(gamma_beta_list) >= 2:
            gamma_beta_list, energy_list = zip(*sorted(zip(gamma_beta_list, energy_list)))
            ax.plot(
                gamma_beta_list, energy_list,
                color=style['color'],
                lw=1.5,
                alpha=style['alpha'],
                zorder=style['zorder'] - 2
            )

    ax.axvline(x=1, color='black', linestyle='--', linewidth=2, zorder=0)
    ax.text(0.35, 0.95, "Non-relativistic", transform=ax.transAxes, fontsize=12, fontweight="bold")
    ax.text(0.65, 0.95, "Relativistic", transform=ax.transAxes, fontsize=12, fontweight="bold")

    # Legend
    label_order = ['ASKAP J0055-2558_c5', 'ASKAP J0055-2558_c15', 'ASKAP J0055-2558_c30', 'ASKAP J0055-2558_sph']
    handles = [mpatches.Patch(color=plot_style[label]['color'], label=label)
               for label in label_order if label in used_labels]
    ax.legend(handles=handles, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2, frameon=False, fontsize=11)

    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel(r'$\Gamma \beta$', fontsize=16)
    ax.set_ylabel(r'Kinetic Energy (erg)', fontsize=16)

    ax.tick_params(axis='both', colors='black', direction='in', length=7, width=1, labelsize="large")
    ax.set_yscale('log')

    from matplotlib.ticker import LogLocator
    ax.yaxis.set_minor_locator(LogLocator(base=10.0, subs=np.arange(2, 10) * 0.1, numticks=10))
    ax.yaxis.set_minor_formatter(plt.NullFormatter())

    for spine in ax.spines.values():
        spine.set_linewidth(2)

    plt.tight_layout()
    return fig

# --- Run the function and show the plot ---
askap_types = [
    'ASKAP J0055-2558_sph',
    'ASKAP J0055-2558_c15',
    'ASKAP J0055-2558_c30',
    'ASKAP J0055-2558_c5',
]
askap_data = all_data[all_data['Type'].isin(askap_types)]

# fig = plot_energy_vs_gamma_beta(askap_data)
from matplotlib.cm import tab10

fig, ax = plt.subplots(figsize=(7, 6))
colors = tab10.colors

# Plot points and fits per type
for i, obj_type in enumerate(askap_types):
    sub_data = askap_data[askap_data['Type'] == obj_type]
    x_vals, y_vals = [], []

    for _, row in sub_data.iterrows():
        gamma_beta_raw = str(row['Gamma_beta'])
        energy_raw = str(row['Energy'])

        if '<' in gamma_beta_raw or '>' in gamma_beta_raw or '<' in energy_raw or '>' in energy_raw:
            continue

        try:
            gamma_beta_val, _, _ = parse_value_with_errors(gamma_beta_raw)
            energy_val, _, _ = parse_value_with_errors(energy_raw)
            if gamma_beta_val > 0 and energy_val > 0:
                x_vals.append(np.log10(gamma_beta_val))
                y_vals.append(np.log10(energy_val))
        except:
            continue

    if len(x_vals) >= 2:
        x_vals = np.array(x_vals)
        y_vals = np.array(y_vals)
        (slope, intercept), cov = np.polyfit(x_vals, y_vals, deg=1, cov=True)
        slope_err = np.sqrt(cov[0, 0])

        # Print fit parameters
        print(f"{obj_type}: log10(E) = ({slope:.2f} Â± {slope_err:.2f}) Ã— log10(Î“Î²) + {intercept:.2f}")

        # Plot data points
        style = plot_style.get(obj_type, {'color': colors[i % len(colors)], 'marker': 'o'})
        gamma_beta_lin = 10**x_vals
        energy_lin = 10**y_vals

        ax.scatter(gamma_beta_lin, energy_lin, label=obj_type, color=style['color'],
                   s=style.get('size', 60), alpha=0.8, edgecolor="black", marker=style['marker'])

        # Plot fit line
        x_fit = np.logspace(min(x_vals) - 0.2, max(x_vals) + 0.2, 200)
        y_fit = 10**(slope * np.log10(x_fit) + intercept)
        ax.plot(x_fit, y_fit, linestyle='--', color=style['color'], alpha=0.8)

# Formatting
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel(r'$\Gamma \beta$', fontsize=16)
ax.set_ylabel(r'Kinetic Energy (erg)', fontsize=16)
ax.set_title("Energy vs Î“Î² with Fits per Geometry", fontsize=14)
ax.legend(fontsize=10)
ax.grid(True, which='both', ls='--', alpha=0.4)

plt.tight_layout()
plt.show()

```
